#Passo a passo de solução do Desafio

```python
#lógica de programação
import os
import pandas as pd

lista_arquivo = os.listdir("/content/drive/MyDrive/Cursos Python")
print(lista_arquivo)
```

```python
tabela_total = pd.DataFrame()

for arquivo in lista_arquivo:
  # se tem "Vendas" no nome do arquivo, então importa o arquivo
  if "Vendas" in arquivo:
    print(f"/content/drive/MyDrive/Cursos Python/{arquivo}")

    tabela = pd.read_excel(f"/content/drive/MyDrive/Cursos Python/{arquivo}")
    tabela_total = pd.concat([tabela_total, tabela], ignore_index=True)

  else:
    print("Arquivo de devolução")

display(tabela_total)
```

```python
# Passo 4 - calcular o produto mais vendido (quantidade)

tabela_produtos = tabela_total.groupby("Produto")['Quantidade'].sum().sort_values(ascending=False)
display(tabela_produtos)
```

```python
# Passo 5 - calcular o produto que mais faturou (em faturamento)

tabela_total['Faturamento'] = tabela_total['Quantidade'] * tabela_total['Valor Unitário']
display(tabela_total)

# agrupa pelo "Faturamento" em ordem descrescente
tabela_faturamento = tabela_total.groupby("Produto")['Faturamento'].sum().sort_values(ascending=False)
display(tabela_faturamento)

```

```python
# Passo 6 - Calcular a loja/cidade que mais vendeu ( em faturamento) -- criar um gráfico/dashboard

tabela_lojas = tabela_total.groupby('ID Loja')['Faturamento'].sum().sort_values(ascending=False)
display(tabela_lojas)
```

```python
import plotly.express as px

grafico = px.bar(tabela_lojas, x= tabela_lojas.index , y='Faturamento')
grafico.show()
```
